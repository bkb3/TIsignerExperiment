```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.6)

import functions

def access(seq, utr):
    new = functions.Optimiser(seq=seq, utr=utr)
    return new.accessibility(new_seq=seq)
    
def expression_score(seq, utr):
    post_prob = functions.get_prob_pos(access(seq, utr))
    return functions.scaled_prob(post_prob)


# from ipywidgets import interact, interactive, fixed, interact_manual
# import ipywidgets as widgets
```

```python
rluc = pd.read_excel('Sequences optimized by TIsigner.xlsx', sheet_name='RLuc')
rluc = rluc[~rluc['TIsigner ID'].isna()].reset_index(drop=True)
rluc
```

```python
rluc = rluc[['TIsigner ID', 'First 30 nt', ]].copy()
```

```python
### 
idx = pd.read_csv('Rluc_names.csv', header=None, skiprows=1)
idx = idx[~idx[0].isna()].reset_index(drop=True)
idx.columns = ['pVD', 'TIsigner ID', 'Type']


def getID(x):
    try:
        return  x.split('/')[1].split('▶:')[0]
    except Exception:
        return x

    
    
def getType(x):
    if x == 'TIsigner variant' or x == '“':
        return 'Optimised'
    if 'Native'  in x:
        return 'Native'
    else:
        return 'Commercial'      
    
    
def getCompany(x):
    if getType(x) == 'Commercial':
        return x.split(' ')[0]
    else:
        return np.nan
    
  
    
idx['TIsigner ID'] = idx['TIsigner ID'].apply(getID)

idx['Company'] = idx['Type'].apply(getCompany)
idx['Type'] = idx['Type'].apply(getType)
idx
```

```python
temp_df = rluc.merge(idx, on='TIsigner ID')
temp_df
```

```python
new_df = pd.concat([temp_df, idx.loc[10:]])
new_df
```

```python
new_df['First 30 nt'][10] = native[:30]
new_df['First 30 nt'][11] = geneart[:30]
new_df['First 30 nt'][12] = idt[:30]
new_df['First 30 nt'][13] = epoch[:30]
new_df['First 30 nt'][14] = genscript[:30]
new_df['First 30 nt'][15] = atum[:30]
new_df
```

```python

new_df['Full Sequence'] = new_df['First 30 nt'].apply(lambda x: x+native[30:])
new_df['Full Sequence'][10] = native
new_df['Full Sequence'][11] = geneart
new_df['Full Sequence'][12] = idt
new_df['Full Sequence'][13] = epoch
new_df['Full Sequence'][14] = genscript
new_df['Full Sequence'][15] = atum
```

```python
new_df['Opening Energy'] = new_df['Full Sequence'].apply(lambda x: access(x, utr))
new_df['Opening Energy'] = new_df['Opening Energy'].round(2)
new_df['Expression Score'] = new_df['Full Sequence'].apply(lambda x: expression_score(x, utr))
new_df['Expression Score'] = new_df['Expression Score'].round(2)
new_df
```

```python
new_df.to_csv('RLuc/RLuc_Index.csv', index=None)
```

#### Making random data for website test

```python
times = []
for i in range(8):
    
    for j in range(3):
        times.append('{}h {}min'.format(i, j*20))
```

```python
import numpy as np
np.random.seed(12345)

fl_df = pd.DataFrame(index=new_df.index, columns=times)
for i in times:
    rnd_fls = np.round(np.random.uniform(low=10, high=50, size=[fl_df.shape[0], 5]), 2)
    fl_df[i] = rnd_fls.tolist()
fl_df = fl_df.round(2)
```

```python
fl_df
```

```python
std_df = pd.DataFrame(index=new_df.index, columns=times)
mean_df = pd.DataFrame(index=new_df.index, columns=times)

for i in times:
    std_df[i] = fl_df[i].apply(lambda x: np.std(x, ddof=1))
    mean_df[i] = fl_df[i].apply(np.mean)
    
std_df = std_df.round(2)
mean_df = mean_df.round(2)


all_dict = {}

for i, v in enumerate(new_df.index):
    new_dict = {}
    new_dict['Type'] = new_df.loc[i]['Type']
    new_dict['Company'] = new_df.loc[i]['Company']
    new_dict['First 30 nt'] = new_df.loc[i]['First 30 nt']
    new_dict['Opening Energy'] = new_df.loc[i]['Opening Energy']
    new_dict['Expression Score'] = new_df.loc[i]['Expression Score']
    data = {}
    for p, q in enumerate(times):
        data_dict = {}
        data_dict["Time"] = q
        data_dict["All Fluorescence"] = fl_df.loc[i][q]
        data_dict["Mean Fluorescence"] = mean_df.loc[i][q]
        data_dict["Std Fluorescence"] = std_df.loc[i][q]
        data[p] = data_dict
    new_dict['Data'] = [v for k, v in data.items()]
    all_dict[i] = new_dict
# all_dict
```
